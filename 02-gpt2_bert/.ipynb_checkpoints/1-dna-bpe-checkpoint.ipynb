### 1. **K-mer åˆ†è§£**

**å®šä¹‰**ï¼šK-mer æ˜¯æŒ‡é•¿åº¦ä¸º k çš„è¿ç»­å­åºåˆ—ã€‚ä¾‹å¦‚ï¼Œåœ¨ DNA åºåˆ—ä¸­ï¼Œä¸€ä¸ª 3-mer å¯èƒ½æ˜¯ "ATG" æˆ– "CGA"ã€‚

**åº”ç”¨**ï¼š
- **DNA åºåˆ—**ï¼šå¸¸ç”¨çš„ k å€¼èŒƒå›´ä» 3 åˆ° 6ã€‚è¾ƒå°çš„ k å€¼å¯ä»¥æ•æ‰åˆ°æ›´ç»†ç²’åº¦çš„ä¿¡æ¯ï¼Œè€Œè¾ƒå¤§çš„ k å€¼åˆ™æœ‰åŠ©äºè¯†åˆ«æ›´é•¿çš„æ¨¡å¼ã€‚
- **è›‹ç™½è´¨åºåˆ—**ï¼šk å€¼é€šå¸¸è¾ƒå¤§ï¼Œå› ä¸ºæ°¨åŸºé…¸çš„æ•°é‡è¾ƒå¤šï¼ˆ20 ç§ï¼‰ï¼Œè¾ƒé•¿çš„ k-mer å¯ä»¥æ•æ‰åˆ°é‡è¦çš„ç»“æ„åŸŸæˆ–åŠŸèƒ½åŒºåŸŸã€‚

**ä¼˜ç‚¹**ï¼š
- ç®€å•ä¸”ç›´è§‚ï¼Œæ˜“äºå®ç°ã€‚
- å¯ä»¥æ•æ‰åˆ°çŸ­åºåˆ—ä¸­çš„å±€éƒ¨ç‰¹å¾ã€‚

**ç¼ºç‚¹**ï¼š
- å¯¹äºéå¸¸é•¿çš„åºåˆ—ï¼Œç”Ÿæˆçš„ k-mer æ•°é‡ä¼šéå¸¸å¤§ï¼Œå¯¼è‡´ç»´åº¦çˆ†ç‚¸é—®é¢˜ã€‚
- ä¸åŒä½ç½®çš„ k-mer ä¹‹é—´ç¼ºä¹ä¸Šä¸‹æ–‡å…³ç³»ã€‚

```python
#ç¤ºä¾‹ä»£ç ï¼ˆPythonï¼‰

def k_mer(seq, k):
    return [seq[i:i+k] for i in range(len(seq) - k + 1)]

dna_sequence = "ATGCGTACGTA"
protein_sequence = "MKQHKAMIVALIVLITAY"

print("DNA 3-mers:", k_mer(dna_sequence, 3))
print("Protein 4-mers:", k_mer(protein_sequence, 4))
```

### 2. **æ»‘åŠ¨çª—å£**

**å®šä¹‰**ï¼šæ»‘åŠ¨çª—å£æ–¹æ³•é€šè¿‡è®¾å®šä¸€ä¸ªå›ºå®šå¤§å°çš„çª—å£æ²¿ç€åºåˆ—ç§»åŠ¨ï¼Œå¹¶åœ¨æ¯ä¸ªä½ç½®æå–çª—å£å†…çš„å­åºåˆ—ã€‚è¿™ä¸ K-mer ç±»ä¼¼ï¼Œä½†å…è®¸é‡å ã€‚

**åº”ç”¨**ï¼š
- **DNA å’Œè›‹ç™½è´¨åºåˆ—**ï¼šçª—å£å¤§å°å¯ä»¥æ ¹æ®å…·ä½“ä»»åŠ¡è°ƒæ•´ï¼Œå¦‚åŸºå› é¢„æµ‹ã€è›‹ç™½è´¨ç»“æ„é¢„æµ‹ç­‰ã€‚

**ä¼˜ç‚¹**ï¼š
- æä¾›äº†æ›´å¤šçš„çµæ´»æ€§ï¼Œå¯ä»¥æ§åˆ¶çª—å£çš„æ­¥é•¿å’Œå¤§å°ã€‚
- æœ‰åŠ©äºæ•æ‰å±€éƒ¨å’Œå…¨å±€ç‰¹å¾ã€‚

**ç¼ºç‚¹**ï¼š
- è®¡ç®—å¤æ‚åº¦è¾ƒé«˜ï¼Œå°¤å…¶æ˜¯å½“çª—å£å¤§å°è¾ƒå¤§æ—¶ã€‚

```python
def sliding_window(seq, window_size, step=1):
    return [seq[i:i+window_size] for i in range(0, len(seq) - window_size + 1, step)]

dna_sequence = "ATGCGTACGTA"
protein_sequence = "MKQHKAMIVALIVLITAY"

print("Sliding window (DNA, size=3, step=1):", sliding_window(dna_sequence, 3))
print("Sliding window (Protein, size=4, step=2):", sliding_window(protein_sequence, 4, step=2))
```

### 3. **è¯è¡¨åˆ†è¯å’ŒåµŒå…¥å¼è¡¨ç¤º**

**å®šä¹‰**ï¼šä½¿ç”¨é¢„è®­ç»ƒçš„åµŒå…¥æ¨¡å‹ï¼ˆå¦‚ Word2Vecã€BERT ç­‰ï¼‰æ¥å°†æ¯ä¸ª token æ˜ å°„åˆ°é«˜ç»´å‘é‡ç©ºé—´ä¸­ã€‚å¯¹äºç”Ÿç‰©åºåˆ—ï¼Œå¯ä»¥ä½¿ç”¨ä¸“é—¨è®¾è®¡çš„åµŒå…¥æ¨¡å‹ï¼Œå¦‚ ProtTransã€ESM ç­‰ã€‚

**åº”ç”¨**ï¼š
- **DNA å’Œè›‹ç™½è´¨åºåˆ—**ï¼šåµŒå…¥æ¨¡å‹å¯ä»¥æ•æ‰åˆ°åºåˆ—ä¸­çš„è¯­ä¹‰ä¿¡æ¯å’Œä¸Šä¸‹æ–‡ä¾èµ–å…³ç³»ã€‚

**ä¼˜ç‚¹**ï¼š
- æ•æ‰åˆ°ä¸°å¯Œçš„è¯­ä¹‰ä¿¡æ¯ï¼Œé€‚åˆå¤æ‚çš„ä¸‹æ¸¸ä»»åŠ¡ã€‚
- å¯ä»¥åˆ©ç”¨å¤§è§„æ¨¡é¢„è®­ç»ƒæ¨¡å‹çš„ä¼˜åŠ¿ã€‚

**ç¼ºç‚¹**ï¼š
- éœ€è¦å¤§é‡çš„è®¡ç®—èµ„æºæ¥è¿›è¡Œé¢„è®­ç»ƒã€‚
- æ¨¡å‹å¤æ‚åº¦è¾ƒé«˜ï¼Œè§£é‡Šæ€§è¾ƒå·®ã€‚

```python
import subprocess
import os
# è®¾ç½®ç¯å¢ƒå˜é‡, autodlä¸€èˆ¬åŒºåŸŸ
result = subprocess.run('bash -c "source /etc/network_turbo && env | grep proxy"', shell=True, capture_output=True, text=True)
output = result.stdout
for line in output.splitlines():
    if '=' in line:
        var, value = line.split('=', 1)
        os.environ[var] = value

"""
import os

# è®¾ç½®ç¯å¢ƒå˜é‡, autodlä¸“åŒº å…¶ä»–idc
os.environ['HF_ENDPOINT'] = 'https://hf-mirror.com'

# æ‰“å°ç¯å¢ƒå˜é‡ä»¥ç¡®è®¤è®¾ç½®æˆåŠŸ
print(os.environ.get('HF_ENDPOINT'))
"""
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# åŠ è½½é¢„è®­ç»ƒçš„è›‹ç™½è´¨åµŒå…¥æ¨¡å‹
tokenizer = AutoTokenizer.from_pretrained("dnagpt/gpt_dna_v0")
model = AutoModel.from_pretrained("dnagpt/gpt_dna_v0")

dna_sequence = "ATGCGTACGTA"
print(tokenizer.tokenize(dna_sequence))

# ç¼–ç åºåˆ—
inputs = tokenizer(dna_sequence, return_tensors="pt")

# è·å–åµŒå…¥
with torch.no_grad():
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state

print("Embeddings shape:", embeddings.shape)
```

å› æ­¤ï¼Œæˆ‘ä»¬ä¹Ÿæ¼”ç¤ºä¸‹ä»å¤´è®­ç»ƒä¸€ä¸ªDNA BPEåˆ†è¯å™¨çš„æ–¹æ³•ã€‚

æˆ‘ä»¬é¦–å…ˆçœ‹ä¸‹GPT2æ¨¡å‹ï¼Œé»˜è®¤çš„åˆ†è¯å™¨ï¼Œå¯¹DNAåºåˆ—åˆ†è¯çš„ç»“æœï¼š

```python
from tokenizers import (
    decoders,
    models,
    normalizers,
    pre_tokenizers,
    processors,
    trainers,
    Tokenizer,
)
from transformers import AutoTokenizer
```

```python
gpt2_tokenizer = AutoTokenizer.from_pretrained('gpt2')
gpt2_tokenizer.pad_token = gpt2_tokenizer.eos_token
```

```python
gpt2_tokenizer.tokenize("TGGCGTGAACCCGGGATCGGG")
```

å¯ä»¥çœ‹åˆ°ï¼Œgpt2æ¨¡å‹å› ä¸ºæ˜¯ä»¥è‹±æ–‡ä¸ºä¸»çš„BPEåˆ†è¯æ¨¡å‹ï¼Œåˆ†è§£çš„éƒ½æ˜¯1åˆ°2ä¸ªå­—æ¯çš„ç»“æœï¼Œè¿™æ ·æ˜¾ç„¶å¾ˆéš¾å……åˆ†è¡¨è¾¾ç”Ÿç‰©è¯­ä¹‰ï¼Œå› æ­¤ï¼Œæˆ‘ä»¬ä½¿ç”¨DNAåºåˆ—æ¥è®­ç»ƒ1ä¸ªBPEåˆ†è¯å™¨ï¼Œä»£ç ä¹Ÿéå¸¸ç®€å•ï¼š

```python
tokenizer = Tokenizer(models.BPE())
tokenizer.pre_tokenizer = pre_tokenizers.ByteLevel(add_prefix_space=False, use_regex=False) #use_regex=False,ç©ºæ ¼å½“æˆä¸€èˆ¬å­—ç¬¦ä¸²
trainer = trainers.BpeTrainer(vocab_size=30000, special_tokens=["<|endoftext|>"]) #3w words
```

```python
tokenizer.train(["../01-data_env/data/dna_1g.txt"], trainer=trainer) #all file list, take 10-20 min
```

```python
encoding = tokenizer.encode("TGGCGTGAACCCGGGATCGGG")
print(encoding.tokens)
```

å¯ä»¥çœ‹åˆ°ï¼Œä»¥DNAæ•°æ®è®­ç»ƒçš„åˆ†è¯å™¨ï¼Œåˆ†è¯æ•ˆæœæ˜æ˜¾è¦å¥½çš„å¤šï¼Œå„ç§é•¿åº¦çš„è¯éƒ½æœ‰ã€‚

```python
tokenizer.save("dna_bpe_dict.json")
```

```python
#ç„¶åæˆ‘ä»¬å¯ä»¥ä½¿ç”¨from_file() æ–¹æ³•ä»è¯¥æ–‡ä»¶é‡Œé‡æ–°åŠ è½½ Tokenizer å¯¹è±¡ï¼š
new_tokenizer = Tokenizer.from_file("dna_bpe_dict.json")
```

```python
#è¦åœ¨ ğŸ¤— Transformers ä¸­ä½¿ç”¨è¿™ä¸ªæ ‡è®°å™¨ï¼Œæˆ‘ä»¬å¿…é¡»å°†å®ƒåŒ…è£¹åœ¨ä¸€ä¸ª PreTrainedTokenizerFast ç±»ä¸­
from transformers import GPT2TokenizerFast
dna_tokenizer = GPT2TokenizerFast(tokenizer_object=new_tokenizer)
dna_tokenizer.save_pretrained("dna_bpe_dict")
#dna_tokenizer.push_to_hub("dna_bpe_dict_1g", organization="dnagpt", use_auth_token="hf_*****") # push to huggingface
```

```python
tokenizer_new = AutoTokenizer.from_pretrained('dna_bpe_dict')
```
